In this lab we will explore how to create effective visualizations using widely available Python libraries. Our focus will be on conveying information that may be relevant to a statistical or scientific problem. We will start off by using the CDC data set from Lab 1 again.

```python
import pandas as pd # library for basic data manipulation tools
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read in data
cdc_df = pd.read_csv("./cdc.csv")

# correct data types for categorical variables
cdc_df = cdc_df.astype(dict(genhlth = "category",
                            exerany = "category",
                            hlthplan = "category",
                            smoke100 = "category",
                            gender = "category"
                           )
                      )
```

## Univariate summaries - continuous

### Histogram

```python
plt.hist(cdc_df.age)
plt.title("Age of Respondents")
plt.xlabel("Age (years)")
plt.ylabel("Count")
pass # excludes pesky arrays that also print
```

```python
sns.kdeplot(cdc_df.age, shade = True, color = "lightgreen")
plt.title("Density Estimate for Age")
plt.xlabel("Age (years)")
plt.ylabel("Density")
pass
```

### Boxplot

```python
plt.boxplot(cdc_df.weight)
plt.title("Weight of CDC BRFSS Respondents")
plt.ylabel("Weight (pounds)")
pass
```

Matplotlib may not have the best functionality for boxplots - we usually want to visually compare subgroups as part of our exploratory data analysis.

```python
sns.boxplot(x = cdc_df.smoke100, y = cdc_df.weight)
plt.title("Weight of Respondents Grouped by Cigarette Use")
plt.ylabel("Weight (pounds)")
plt.xlabel("Respondent Smoked 100 or more Cigarettes in Lifetime")
pass
```

### Bivariate visualizations - the scatterplot

```python
plt.scatter(x = cdc_df.height, y = cdc_df.weight)
plt.title("Height vs. Weight for CDC BRFSS Respondents")
plt.xlabel("Height (inches)")
plt.ylabel("Weight (pounds)")
pass
```

## Summaries of discrete quantities

```python
sns.catplot(x = "genhlth", 
            data = cdc_df, 
            kind = "count", 
            color = "lightblue",
            order = ("excellent", "very good", "good", "fair", "poor")
           )
pass
```

## Facet wrap and multi-window plots

```python
g1 = sns.FacetGrid(data = cdc_df, col = "genhlth") # set grid
g1.map(plt.hist, "age") # tell seaborn what to plot on grid
g1.axes[0,0].set_ylabel("Count") # labels only first plot, looks nicer
# g1.set(ylabel = "Count") blunt instrument - labels all
pass
```

```python
cdc_df.groupby(["genhlth"]).mean() # get summary of numeric variables by a categorical grouping
```
